# Figure 2 and 4
This code reproduces the figures 2 and 4 of our report  the paper  *Centers for Disease Control and Prevention as a Strategic Agent in the Pediatric Vaccine Market: An Analytical Approach*.


```python
#Import required packages and define useful functions
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FuncFormatter
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

#Function for fomatting the axes of the graph of Figure 4
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.0f' % (x * 1e-6)

#Paramenter calculation for figure 4
def parameter_calculation(gamma):
    
    '''
    Compute the parameters that depende on gamma.
    '''
    
    p_priv = (34.190+10.3*gamma)*((1-gamma)/(2-gamma)) #private sector price, equal for each manufacturer, eq. 20
    
    q_priv = 1e6*(3.419 + 1.03*gamma)/\
             ((1 + gamma)*(2 - gamma)) #private sector quantity, equal for each manufacturer, eq. 21
    
    U = 1e6*((3.419 + 1.03*gamma)/gamma)*(1-(2*(1-gamma)**0.5)/\
        (((1+gamma)**0.5)*(2-gamma))) #private-sector capacity lower bound, eq. 22
    
    alpha = 1553000+1365000*(1+gamma) #obtained from eq. 12 of article and eq. 3 of sumplement
    
    return p_priv, q_priv, U, alpha
```

## Figure 2

```python
#Get values of gamma
gamma = np.linspace(0.001, 1, 100)

#_____________________________________________Figure 2a)_________________________________________________
#Compute parameter a_priv and a_pub
a_priv = (1.03 + 2.389/(1 + gamma)) #Eq. 7
a_pub = (1.365 + 1.553/(1 + gamma)) #Eq. 9

plt.plot(gamma, a_priv, label="Private", linewidth=2, color='black')
plt.xlim([0,1])
plt.ylim([0,4])

plt.plot(gamma, a_pub,'--',color='black',label = "Public", linewidth=2)
plt.legend(loc='lower left', shadow=False, fontsize='large', title = 'Sector',title_fontsize = 'x-large')
plt.ylabel('Intercept (million units)',fontsize='large'); plt.xlabel('Product similarity',fontsize='large')

plt.show()

#_____________________________________________Figure 2b)_________________________________________________
p_priv = (34.19 + 10.3*gamma)*(1-gamma)/(2-gamma) #Private price, eq. 9

plt.plot(gamma, p_priv, color= 'black')
plt.xlim([0,1]); plt.ylim([0,20])
plt.xlabel('Product similarity', fontsize = 'large'); plt.ylabel('Price (USD)', fontsize = 'large'); 

plt.show()

#_____________________________________________Figure 2c)_________________________________________________
q_priv = (3.419 + 1.030*gamma)/((1+gamma)*(2-gamma)) #Private quantity, eq. 10

U = ((3.419 + 1.03*gamma)/gamma)*(1-(2*(1-gamma)**0.5)/\
    (((1+gamma)**0.5)*(2-gamma))) #private-sector capacity lower bound, eq. 11

plt.plot(gamma, q_priv, lw=2, color = 'black', label= 'Quantity')
plt.plot(gamma, U,'--',lw=2, color = 'black',label = 'Capacity')
plt.xlim([0,1])
plt.ylim([0,4])
plt.xlabel('Product similarity', fontsize = 'large'); plt.ylabel('Quantity (million units)', fontsize = 'large'); 
plt.legend(loc='lower left', shadow=False, fontsize='large', title = 'Parameter',title_fontsize = 'x-large')

plt.show()
```

## Figure 4
The following code uses an analtycal approach to draw the feasible region for the following problem conditions:

- a) Base case scenario.
- b) Symmetric scenario, both manufacturers with the same target profit and product similarity $\gamma$ = 0.25.
- c) Symmetric scenario, both manufacturers with the same Target profit and higher product similarity ($\gamma$ = 0.5).

```python
#The following data is common for all graphs

rho_inf = 18.62  #Inflation bond infanrix
rho_dap = 18.02  #Inflation bond daptacel
q_pubm = np.linspace(0.01, 4e6, 100) #capacity of manufacturer

#_____________________________________________Figure 4 a)__________________________________________________________

gamma = 0.25
p_priv, q_priv, U, alpha =parameter_calculation(gamma)
Pinf = 39.8e6    #Target profit infanrix
Pdap = 45.1e6    #Target profit daptacel

q_inf = (1e5*(-p_priv*q_priv + Pdap)+(q_pubm**2)-q_pubm*alpha)/(-q_pubm*gamma) #Eq. 5 and 1e
q_dap = (1e5*(-p_priv*q_priv + Pinf)+(q_pubm**2)-q_pubm*alpha)/(-q_pubm*gamma) #Eq. 5 and 1e

fig, ax1 = plt.subplots(figsize=(6, 6))

#Target profit constraint
ax1.plot(q_pubm ,q_inf, color ='k')
ax1.plot(q_dap, q_pubm, color ='k')


#Non negativity constraint
ax1.plot(q_pubm, (alpha-gamma*q_pubm), color ='k',linestyle=':') #Infanrix price - 'horizontal', eq. 5 and 1i
ax1.plot((alpha-gamma*q_pubm), q_pubm, color ='k',linestyle=':') #Daptacel price - 'vertical', eq. 5 and 1i

#inflation limit
bond_inf = -rho_inf*1e5+alpha-gamma*q_pubm  #price bond infanrix, eq. 5 and 1h
bond_dap = -rho_dap*1e5+alpha-gamma*q_pubm  #price bond daptacel, eq. 5 and 1h
ax1.plot(q_pubm, bond_inf, color ='b',linestyle=':') 
ax1.plot(bond_dap, q_pubm, color ='b',linestyle=':')

#Public sector capacity
#If one manufacture offers 0 capacity to public sector, the other will supply all the public demand
x1,y1=(0,4034000*0.57)
#If one manufacturer offers all its capacity to public sector, the other can supply to public sector its capacity
#minus the private sector demand
x2,y2= (4034000,4034000-U)
#Get line slope
slope= (y2-y1)/(x2-x1)
U_eq = slope*q_pubm-slope*x2+y2 #eq. 1g
ax1.plot(q_pubm, U_eq, color ='k',linestyle='-.') #Infanrix
ax1.plot(U_eq, q_pubm, color ='k',linestyle='-.') #Daptacel

#Public sector demand (eq. 1c)
ax1.plot(q_pubm, 4.034e6*0.57-q_pubm, '--', color ='k')

#Formatting

ax1.fill_between(q_pubm, q_inf, facecolor = 'lightgray')
ax1.fill_between(bond_dap, q_pubm, facecolor = 'w')
ax1.fill_between(q_pubm, bond_inf, facecolor = 'w')

ax1.set_xlim([0,4e6]), ax1.set_ylim([0,4e6])

formatter = FuncFormatter(millions)

ax1.xaxis.set_major_formatter(formatter)
ax1.yaxis.set_major_formatter(formatter)
ax1.set_xticks(np.arange(0, 5e6, step=1e6))
ax1.set_yticks(np.arange(0, 5e6, step=1e6))
plt.gca().set_aspect('equal', adjustable='box')
ax1.set_xlabel('SP (Daptacel) Public Sector Quantity (millions)', fontsize = 'large')
ax1.set_ylabel('GSK (Infanrix) Public Sector Quantity (millions)', fontsize = 'large')

#Optimal solution
ax1.scatter(2202070, 846732, marker = 'X', s=200,color = 'red')
plt.show()

#_____________________________________________Figure 4 b)__________________________________________________________

gamma = 0.25
p_priv, q_priv, U, alpha =parameter_calculation(gamma)
Pinf = 42.45e6    #Target profit infanrix
Pdap = 42.45e6    #Target profit daptacel

q_inf = (1e5*(-p_priv*q_priv + Pdap)+(q_pubm**2)-q_pubm*alpha)/(-q_pubm*gamma) #Eq. 5 and 1e
q_dap = (1e5*(-p_priv*q_priv + Pinf)+(q_pubm**2)-q_pubm*alpha)/(-q_pubm*gamma) #Eq. 5 and 1e

fig, ax2 = plt.subplots(figsize=(6, 6))

#Target profit constraint
ax2.plot(q_pubm ,q_inf, color ='k')
ax2.plot(q_dap, q_pubm, color ='k')


#Non negativity constraint
ax2.plot(q_pubm, (alpha-gamma*q_pubm), color ='k',linestyle=':') #Infanrix price - 'horizontal', eq. 5 and 1i
ax2.plot((alpha-gamma*q_pubm), q_pubm, color ='k',linestyle=':') #Daptacel price - 'vertical', eq. 5 and 1i

#inflation limit
bond_inf = -rho_inf*1e5+alpha-gamma*q_pubm  #price bond infanrix, eq. 5 and 1h
bond_dap = -rho_dap*1e5+alpha-gamma*q_pubm  #price bond daptacel, eq. 5 and 1h
ax2.plot(q_pubm, bond_inf, color ='b',linestyle=':') 
ax2.plot(bond_dap, q_pubm, color ='b',linestyle=':')


#Public sector capacity
#If one manufacture offers 0 capacity to public sector, the other will supply all the public demand
x1,y1=(0,4034000*0.57)
#If one manufacturer offers all its capacity to public sector, the other can supply to public sector its capacity
#minus the private sector demand
x2,y2= (4034000,4034000-U)
#Get line slope
slope= (y2-y1)/(x2-x1)
U_eq = slope*q_pubm-slope*x2+y2 #eq. 1g
ax2.plot(q_pubm, U_eq, color ='k',linestyle='-.') #Infanrix
ax2.plot(U_eq, q_pubm, color ='k',linestyle='-.') #Daptacel


#Public sector demand (eq. 1c)
ax2.plot(q_pubm, 4.034e6*0.57-q_pubm, '--', color ='k')


#Formatting
ax2.fill_between(q_pubm, q_inf, facecolor = 'lightgray')
ax2.fill_between(bond_dap, q_pubm, facecolor = 'w')
ax2.fill_between(q_pubm, bond_inf, facecolor = 'w')
ax2.fill_between(q_dap, q_pubm, facecolor = 'w', where = (q_pubm < 1.2e6))
ax2.fill_between(q_dap, q_pubm, 4e6, facecolor = 'w', where = (q_pubm > q_dap))
ax2.fill_between(q_pubm, U_eq, 4e6, facecolor = 'w')
ax2.fill_between(U_eq, q_pubm, 4e6, facecolor = 'w')

ax2.set_xlim([0,4e6]), ax2.set_ylim([0,4e6])

ax2.xaxis.set_major_formatter(formatter)
ax2.yaxis.set_major_formatter(formatter)
ax2.set_xticks(np.arange(0, 5e6, step=1e6))
ax2.set_yticks(np.arange(0, 5e6, step=1e6))
plt.gca().set_aspect('equal', adjustable='box')
ax2.set_xlabel('SP (Daptacel) Public Sector Quantity (millions)', fontsize = 'large')
ax2.set_ylabel('GSK (Infanrix) Public Sector Quantity (millions)', fontsize = 'large')


#Feasible solution
ax2.scatter(1950350, 1950350, marker = 'X', s=200,color = 'red')

plt.show()

#_____________________________________________Figure 4 c)__________________________________________________________
gamma = 0.5
p_priv, q_priv, U, alpha =parameter_calculation(gamma)
Pinf = 42.45e6    #Target profit infanrix
Pdap = 42.45e6    #Target profit daptacel

q_inf = (1e5*(-p_priv*q_priv + Pdap)+(q_pubm**2)-q_pubm*alpha)/(-q_pubm*gamma) #Eq. 5 and 1e
q_dap = (1e5*(-p_priv*q_priv + Pinf)+(q_pubm**2)-q_pubm*alpha)/(-q_pubm*gamma) #Eq. 5 and 1e

fig, ax3 = plt.subplots(figsize=(6, 6))

#Target profit constraint
ax3.plot(q_pubm ,q_inf, color ='k', label = 'Target Profit (1e)')
ax3.plot(q_dap, q_pubm, color ='k')


#Non negativity constraint
ax3.plot(q_pubm, (alpha-gamma*q_pubm), color ='k',linestyle=':', label = 'Non-negativity (1i)') #Infanrix price - 'horizontal', eq. 5 and 1i
ax3.plot((alpha-gamma*q_pubm), q_pubm, color ='k',linestyle=':') #Daptacel price - 'vertical', eq. 5 and 1i

#inflation limit
bond_inf = -rho_inf*1e5+alpha-gamma*q_pubm  #price bond infanrix, eq. 5 and 1h
bond_dap = -rho_dap*1e5+alpha-gamma*q_pubm  #price bond daptacel, eq. 5 and 1h
ax3.plot(q_pubm, bond_inf, color ='b',linestyle=':', label = 'Inflation bound (1h)') 
ax3.plot(bond_dap, q_pubm, color ='b',linestyle=':')


#Public sector capacity
#If one manufacture offers 0 capacity to public sector, the other will supply all the public demand
x1,y1=(0,4034000*0.57)
#If one manufacturer offers all its capacity to public sector, the other can supply to public sector its capacity
#minus the private sector demand
x2,y2= (4034000,4034000-U)
#Get line slope
slope= (y2-y1)/(x2-x1)
U_eq = slope*q_pubm-slope*x2+y2 #eq. 1g
ax3.plot(q_pubm, U_eq, color ='k',linestyle='-.', label='Public Sector Capacity (1g)') #Infanrix
ax3.plot(U_eq, q_pubm, color ='k',linestyle='-.') #Daptacel


#Public sector demand (eq. 1c)
ax3.plot(q_pubm, 4.034e6*0.57-q_pubm, '--', color ='k', label='Public Sector Demand (1c)')


#Formatting
ax3.fill_between(q_dap, q_pubm, facecolor = 'lightgray')
ax3.fill_between(q_pubm, q_inf, 4e6, facecolor = 'w')
ax3.fill_between(bond_dap, q_pubm, facecolor = 'w')
ax3.fill_between(q_pubm, bond_inf, facecolor = 'w')

ax3.set_xlim([0,4e6]), ax3.set_ylim([0,4e6])

ax3.xaxis.set_major_formatter(formatter)
ax3.yaxis.set_major_formatter(formatter)
ax3.set_xticks(np.arange(0, 5e6, step=1e6))
ax3.set_yticks(np.arange(0, 5e6, step=1e6))
plt.gca().set_aspect('equal', adjustable='box')
ax3.set_xlabel('SP (Daptacel) Public Sector Quantity (millions)', fontsize = 'large')
ax3.set_ylabel('GSK (Infanrix) Public Sector Quantity (millions)', fontsize = 'large')


#Optimal solution
ax3.scatter(1568400, 1568400, marker = 'X', s=200,color = 'red', label = 'Optimal Solution')
plt.legend()
plt.show()
```
